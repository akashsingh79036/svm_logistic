```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import joblib
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
data.target#species is the target column it's data have three classes setosa,viriginca and versicolor stored as a label encoder
```

```python
df['target']=data.target
```

```python
df
```

```python
data .target_names
```

```python
df['target_names']=df['target'].map({i:name for i,name in enumerate(data.target_names)})
```

```python
df['target_names']
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['target_names'].value_counts()
```

```python
binary_df=df[df['target']!=2]
```

```python
binary_df
```

```python
x_binary=binary_df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
```

```python
y_binary=binary_df[['target']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_binary,y_binary, test_size=0.2)
```

```python
x_train.describe()
```

```python
x_binary.describe()
```

```python
scaler=StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
svm_binary=SVC(kernel='linear',C=1.0,probability=True)
```

```python
svm_binary.fit(x_train,y_train)
```

```python
svm_binary.predict(x_test)
```

```python
print("binary class classification report:\n",classification_report(y_test,svm_binary.predict(x_test)))
```

```python
confusion_matrix(y_test,svm_binary.predict(x_test))
```

```python
joblib.dump(svm_binary,'svm_binary.pkl')
```
